```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import scanpy as sc
from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
Image.MAX_IMAGE_PIXELS = None
```

```python
tag = 'B1'

landmark = pd.read_csv('/home/wcy/code/datasets/SMA/V11T17-102-'+tag+'/V11T17-102_'+tag+'_MSI/landmarks_noHE.csv')
src = landmark[['X_left', 'Y_left']].values
dst = landmark[['X_right', 'Y_right']].values
trans_mtx, inliers = cv2.estimateAffine2D(src, dst)

msi_metadata = pd.read_csv('/home/wcy/code/datasets/SMA/V11T17-102-'+tag+'/V11T17-102_'+tag+'_MSI/MSI_metadata.csv', index_col=0)
msi_intensities = pd.read_csv('/home/wcy/code/datasets/SMA/V11T17-102-'+tag+'/V11T17-102_'+tag+'_MSI/MSI_intensities.csv', index_col=0)

points_homo = np.hstack([msi_metadata.values, np.ones((msi_metadata.shape[0], 1))])  
transformed = points_homo @ trans_mtx.T

img_path = '/home/wcy/code/datasets/SMA/V11T17-102-'+tag+'/V11T17-102_'+tag+'_RNA/outs/spatial/tissue_hires_image.png'
image = Image.open(img_path)
image_data = np.array(image)

plt.imshow(image_data)
plt.scatter(transformed[:, 0], transformed[:, 1], c=msi_intensities['674.28592'], s=5, alpha=0.5)
plt.gca().set_aspect(1)
plt.axis('off')
plt.savefig('/home/wcy/code/datasets/SMA/V11T17-102-'+tag+'/V11T17-102_'+tag+'_RNA/outs/spatial/test_align_msi.png', 
            dpi=600, bbox_inches='tight')
plt.show()
```

```python
adata = sc.read_h5ad('/home/wcy/code/datasets/SMA/V11T17-102-B1/V11T17-102_B1_MSI/msi_adata.h5ad')
scale = adata.uns['spatial']['V11T17-102_B1']['scalefactors']['tissue_hires_scalef']
spatial = transformed/scale
adata.obsm['spatial'] = spatial
adata.write_h5ad('/home/wcy/code/datasets/SMA/V11T17-102-B1/V11T17-102_B1_MSI/msi_adata.h5ad')
```
